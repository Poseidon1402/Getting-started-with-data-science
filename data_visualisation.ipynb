```python
from matplotlib import pyplot as plt
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# draw a plot ;  years in horizontal axes, GDP (PIB) on vertical axes
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# Add title
plt.title('Valeurs de du PIB am√©ricain')

# Add label on the y axes
plt.ylabel('Milliards de dollars')
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# Draw bars with:
#   * X-axis [nb_oscars]
#   * y-axis [0, 1, 2, 3, 4] (films'indexes)
plt.bar(range(len(movies)), num_oscars)
plt.title('Mes films pr√©f√©r√©s')
plt.ylabel('Nombre d\' Oscars')

#label on x-axis with the names of movies in the middle of the bars
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# categorizes scores by deciles, but places 100 in the 90 category
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()], # bars shifted by 5 to the right
    histogram.values(), # gives each bar its correct height
    10, #  height to 10
    edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5]) # x-axis from -5 to 105 and y-axis from 0 to 5
plt.xticks([10 * i for i in range(11)]) # labels in x multiples of 10
plt.xlabel("D√©ciles")
plt.ylabel("Nombre d'√©tudiants")
plt.title("R√©partition des notes d'examen")
plt.show()  
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# We can multiply calls to plt.plot
# To show a plethora of series on the same graphic
# solid green line:
plt.plot(xs, variance, 'g-', label='variance')

# solid red line:
plt.plot(xs, bias_squared, 'r-', label='bias¬≤')

# dashed blue line:
plt.plot(xs, total_error, 'b:', label='total error')

# loc = 9 means top center
plt.legend(loc=9)
plt.xlabel('Complexit√© du mod√®le')
plt.xticks([])

plt.title('Compromis entre biais et variance')
plt.show()
```

```python
# A scatter plot to represent the number of friends compared with time spent on the site
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)

# label for each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        # place the label near the associated point
        xy = (friend_count, minute_count),
        xytext = (5, -5),
        textcoords = 'offset points'
    )
plt.title("Minutes par jour / Nombre d'amis")
plt.xlabel("Nombre d'amis")
plt.ylabel("Temps quotidien sur le site (min)")
plt.show()
```
